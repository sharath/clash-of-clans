```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from clan import generate_clan
```

```python
plt.style.use('seaborn-white')
```

```python
clans = [generate_clan() for i in range(2)]
steps = 10000
dt = 700/steps 
x = np.linspace(0, 700, steps)

for t in x:
    for c in clans:
        c.step(dt)
```

```python
plt.gcf().set_size_inches(8, 6)
sns.despine()
plt.xlabel('Time')
plt.ylabel('Population')
for n, c in enumerate(clans):
    plt.plot(x[:2300], np.add(c.history['m'][:2300],c.history['f'][:2300]), label=f'Clan {n+1}', c=c.color)
plt.legend()
plt.show()
```

```python
plt.gcf().set_size_inches(8, 6)
sns.despine()
plt.xlabel('Time')
plt.ylabel('Ratio Female')
for n, c in enumerate(clans):
    plt.plot(x[:2300], c.history['s'][:2300], label=f'Clan {n+1}', c=c.color)
plt.legend()
plt.show()
```

```python
plt.gcf().set_size_inches(8, 6)
sns.despine()
plt.xlabel('Time')
plt.ylabel('Radius')
for n, c in enumerate(clans):
    plt.plot(x[:2300], c.history['radius'][:2300], label=f'Clan {n+1}', c=c.color)
plt.legend()
plt.show()
```
